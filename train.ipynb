```python
import yaml
import os

with open(data_yaml_path, 'r') as file:
    data = yaml.safe_load(file)
    
data["path"] = "dataset"
data["train"] = "dataset/train/images"
data["test"] = "dataset/test/images"

modified_data_yaml_path = "dataset/modified_data.yaml"
with open(modified_data_yaml_path, 'w') as file:
    yaml.dump(data, file)
    
print(f"Modified data.yaml saved to {modified_data_yaml_path}")

with open(modified_data_yaml_path, 'r') as file:
    print(file.read())
```

```python
from ultralytics import YOLO


model = YOLO("yolo11n.pt")


model.train(
    data="dataset/data.yaml",
    epochs=3,
    imgsz=640,
    batch=16,
    name="seatbelt_detector"
)
```
